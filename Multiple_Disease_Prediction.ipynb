```python
df=pd.read_csv('/content/indian_liver_patient - indian_liver_patient.csv')
df
```

```python
#cateceri=("Gender","Dataset")
#contunius=("Age","Total_Bilirubin","Direct_Bilirubin","Alkaline_Phosphotase","Alamine_Aminotransferase","Aspartate_Aminotransferase","Total_Protiens","Albumin","Albumin_and_Globulin_Ratio")
```

```python
df.corr()
```

```python
df.nunique()
```

```python
for i in df.select_dtypes(include='object').columns:
  print(i,df[i].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
```

```python
df['Gender']=encoder.fit_transform(df[['Gender']])
```

```python
df
```

```python
df=pd.DataFrame(df)
```

```python
# Calculate the A/G ratio using the given formula
df['Albumin_and_Globulin_Ratio'] =df['Albumin_and_Globulin_Ratio'].fillna(df['Albumin'] / (df['Total_Protiens'] - df['Albumin']))
```

```python
df['Dataset']=df['Dataset'].map({1:1,2:0})
```

```python
df.isnull().sum()
```

```python
df.set_index('Age',inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier().fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier().fit(x_train,y_train)
y_pred=model1.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model2=LogisticRegression().fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
model3=XGBClassifier().fit(x_train,y_train)
y_pred=model3.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
model4=AdaBoostClassifier().fit(x_train,y_train)
y_pred=model4.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import SGDClassifier
liver_model6=SGDClassifier().fit(x_train,y_train)
y_pred=liver_model6.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
import streamlit as st
import pickle
import numpy as np

# Load the model
model = pickle.load(open('liver.pkl', 'rb'))

# Streamlit user interface
st.title("Liver Disease Prediction Model")
st.write("Enter the following details to predict liver disease.")

# Input fields for the features (adjust these according to your model's features)
age = st.number_input("Age", min_value=1, max_value=100)
sex = st.selectbox("Sex", ["Male", "Female"])
bili = st.number_input("Total Bilirubin")
alkphos = st.number_input("Alkaline Phosphatase")
sgpt = st.number_input("SGPT")
sgot = st.number_input("SGOT")
albumin = st.number_input("Albumin")
alb_glob_ratio = st.number_input("Albumin/Globulin Ratio")

# Convert categorical input to numeric (assuming 'Male' is 1 and 'Female' is 0)
sex = 1 if sex == "Male" else 0

# Prepare the input data for prediction
input_data = np.array([[age, sex, bili, alkphos, sgpt, sgot, albumin, alb_glob_ratio]])

# Button to make prediction
if st.button("Predict"):
    prediction = model.predict(input_data)

    # Display the result
    if prediction == 1:
        st.write("Prediction: Liver Disease Detected")
    else:
        st.write("Prediction: No Liver Disease Detected")
```
